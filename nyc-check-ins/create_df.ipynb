```python
neibor_rating_price_review_dic = {}
for index, row in yelp.iterrows():
    lat = row['latitude']
    lng = row['longitude']
    dis = {}
    for index_1, row_1 in neighbor.iterrows():
        dis[index_1] = GetDistance(lat, lng, row_1['lat'], row_1['lng'])
    pair = min(dis.items(), key=lambda x: x[1])
    closest_neighbor = pair[0]
    if closest_neighbor not in neibor_rating_price_review_dic:
        neibor_rating_price_review_dic[closest_neighbor] = {}
    else:
        if 'count' not in neibor_rating_price_review_dic[closest_neighbor]:
            try:
                neibor_rating_price_review_dic[closest_neighbor]['count'] = 1
                neibor_rating_price_review_dic[closest_neighbor]['rating'] = row['rating']
                neibor_rating_price_review_dic[closest_neighbor]['price'] = len(row['price'])
                neibor_rating_price_review_dic[closest_neighbor]['review_count'] = row['review count']
            except:
                continue
        else:
            try:
                neibor_rating_price_review_dic[closest_neighbor]['count'] += 1
                neibor_rating_price_review_dic[closest_neighbor]['rating'] += row['rating']
                neibor_rating_price_review_dic[closest_neighbor]['price'] += len(row['price'])
                neibor_rating_price_review_dic[closest_neighbor]['review_count'] += row['review count']
            except:
                continue
        
        
```

```python
avg_dic = {}
for key in neibor_rating_price_review_dic:
    temp = neibor_rating_price_review_dic[key]
    try:
        count = temp['count']
        temp['price'] = temp['price'] / count
        temp['rating'] = temp['rating'] / count
        temp['review_count'] = temp['review_count'] / count
        avg_dic[key] = temp
        temp.pop('count')
    except:
        continue
```

```python
price = []
review_count = []
rating = []
for i in range(100):
    if i in avg_dic:
        price.append(avg_dic[i]['price'])
        review_count.append(avg_dic[i]['review_count'])
        rating.append(avg_dic[i]['rating'])
    else:
        price.append(0)
        review_count.append(0)
        rating.append(0)
```

```python
neighbor['rating'] = rating
neighbor['review_count'] = review_count
neighbor['price'] = price
```

```python
def getNearestAverage(i):
    distance = []
    for index, row in rent.iterrows():
        n_l = neighbor.as_matrix()[i]
        n_lat = n_l[1]
        n_lng = n_l[2]
        distance.append(GetDistance(n_lat, n_lng, row['lat'], row['lng']))
    nearest_index = np.argsort(np.array(distance))[:3]
    price_value = rent.as_matrix()[nearest_index][:, 3].sum()
    return price_value / 3
        
```

```python
neighbor_price = {}
for index, row in rent.iterrows():
    distance = []
    for index_1, row_1 in neighbor.iterrows():
        distance.append(GetDistance(row['lat'], row['lng'], row_1['lat'], row_1['lng']))
        n = np.argsort(np.array(distance))[0]
        if n not in neighbor_price:
            neighbor_price[n] = {}
            neighbor_price[n]['count'] = 1
            neighbor_price[n]['price'] = row['price']
        else:
            neighbor_price[n]['count'] += 1
            neighbor_price[n]['price'] += row['price']
        
# There will be some neighborhoods without data, we calculate average price for nearest 3 point as its price
price_col = []
for i in range(100):
    if i not in neighbor_price:
        avg_price_nearest = getNearestAverage(i)
        neighbor_price[i] = { "count": 1, 'price': avg_price_nearest }
        price_col.append(avg_price_nearest)
    else:
        price_col.append(neighbor_price[i]['price'] / neighbor_price[i]['count'])
```

```python
neighbor['rent'] = price_col
```
