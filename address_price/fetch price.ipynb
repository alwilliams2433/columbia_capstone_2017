```python
import requests
from bs4 import BeautifulSoup

str1 = "http://www.loopnet.com/for-lease/new-york-ny/multiple-property-types/"
str2 = "/?sk=0d5d210288c7d9887fbdd89ea1d7a3bc&view=list"
```

```python
price = open("price_new.csv", "w")
price.write("address")
price.write(",")
price.write("price")
price.write(",")
price.write("zipcode")
price.write("\n")
history = {"result"}
for i in range(1, 25):
    strcur = str1 + str(i) + str2
    response = requests.get(strcur)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    el1 = soup.find_all(itemprop="streetAddress")
    el2 = soup.find_all(class_="listing-price")
    el3 = soup.find_all(itemprop="postalCode")
    
    for j in range(len(el1)):
        add = el1[j].text
        zipcode = el3[j].text
        pri = el2[j].text
        price.write(add)
        price.write(",")
        if pri != "Rent Not Disclosed":
            pri = pri.split("/")[0].rstrip()
            price.write(pri)
        else:
            price.write("NA")
        price.write(",")
        price.write(zipcode)
        price.write("\n")
        print("address is: " + add)
        print("price is: " + pri)
        history.add(add)

price.close()
```

```python
#richie extension code to pull lat long

import pandas as pd
import geopy
from geopy.geocoders import Nominatim
```

```python
data = pd.read_csv("address_price_zip_cleansed.csv")
geolocator = Nominatim()
length = len(data['address'])
data['lat']=0.0
data['lng']=0.0
for i in range(length):
    try: 
        lat = geolocator.geocode(data['address'][i]+str(" New York, NY")).latitude
        lng = geolocator.geocode(data['address'][i]+str(" New York, NY")).longitude
    except KeyError:
        lat = 0.0
        lng = 0.0
    data['lat'][i] = lat
    data['lng'][i] = lng
    

    
```

```python
data.to_csv("withlatlong.csv")
```
