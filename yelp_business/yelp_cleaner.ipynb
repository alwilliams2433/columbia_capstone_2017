```python
import json

f = open("yelp_business_data_50000.txt", "r")
dic = {}
count = 0
for line in f.readlines():
    data = json.loads(line)
    dic[count] = data
    count += 1
output = open("newdata.csv", "w+")

features = ["name", "price", "rating", "zip code", "latitude", "longitude", "review count"]

for feature in features:
    output.write(feature)
    if feature != "review count":
        output.write(",")
    
output.write("\n")
for key in dic:

    try:
        output.write(dic[key]["name"])
    except:
        continue
    output.write(",")
    
#     price
    try:
        output.write(dic[key]["price"])
    except:
        output.write("NA")
    output.write(",")
# #     # rating
    try:
        output.write(str(dic[key]["rating"]))
    except:
        output.write("NA")
    output.write(",")
# #     zip code
    try:
        output.write(str(dic[key]["location"]["zip_code"]))
    except:
        output.write("NA")
    output.write(",")
    
    try:
        output.write(str(dic[key]["coordinates"]["latitude"]))
    except:
        output.write("NA")
    output.write(",")
    try:
        output.write(str(dic[key]["coordinates"]["longitude"]))
    except:
        output.write("NA")
    output.write(",")
#     review count
    try:
        output.write(str(dic[key]["review_count"]))
    except:
        output.write("NA")
        
    #Omit categories for now
#     output.write(",")
# #     categories
#     try:
#         categories = []
#         for tempdic in dic[key]["categories"]:
#             categories.append(tempdic["alias"])

#         output.write(str(categories))
#     except:
#         output.write("NA")

    output.write("\n")

f.close()
output.close()
```
